```python
# Import the necessary functions from the sympy library
from sympy import symbols, integrate

# 1. Define the symbolic variable
# Let's use 'x' as our variable of integration
x = symbols('x')

# 2. Define the function you want to integrate
# Example: f(x) = x^2 + 2x + 1
function_to_integrate = x**2 + 2*x + 1

# 3. Use the integrate() function to find the indefinite integral
# The function takes two arguments: the expression and the variable of integration
integral_result = integrate(function_to_integrate, x)

# 4. Print the result
print(f"La funzione da integrare è: {function_to_integrate}")
print(f"L'integrale indefinito è: {integral_result}")

# --- Esempio 2: Integrale di una funzione trigonometrica ---
print("\n--- Esempio 2: Integrale trigonometrico ---")

# Definiamo una nuova funzione, ad esempio sin(x)
trig_function = 3 * symbols('x')**3 + 4*symbols('x')**2 - 5*symbols('x') + 7

# Calcoliamo l'integrale indefinito
trig_integral = integrate(trig_function, x)

print(f"La funzione da integrare è: {trig_function}")
print(f"L'integrale indefinito è: {trig_integral}")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# 1. Definire il ticker (simbolo) e l'intervallo di tempo
ticker = "AAPL"
end_date = datetime.now()
start_date = end_date - timedelta(days=5*365)  # Ultimi 5 anni

# 2. Scaricare i dati storici
# Il metodo 'download' di yfinance scarica i dati OHLCV (Open, High, Low, Close, Volume)
dati_aapl = yf.download(ticker, start=start_date, end=end_date)

# 3. Creare il grafico
# Vogliamo visualizzare il prezzo di chiusura ('Close')
plt.figure(figsize=(12, 6))  # Imposta le dimensioni del grafico
plt.plot(dati_aapl.index, dati_aapl['Close'], label='Prezzo di chiusura AAPL')

# 4. Personalizzare il grafico
plt.title(f"Prezzo di chiusura di {ticker} negli ultimi 5 anni")
plt.xlabel("Data")
plt.ylabel("Prezzo (USD)")
plt.legend()  # Mostra la legenda
plt.grid(True)  # Aggiunge una griglia per una migliore leggibilità
plt.show()  # Mostra il grafico
```

```python
# 1. Definire il nome del file e il contenuto
nome_file = "saluto.txt"
contenuto = "ciao"

# 2. Aprire il file in modalità scrittura ('w')
# 'with open(...) as f:' è il metodo preferito perché gestisce automaticamente la chiusura del file
try:
    with open(nome_file, "w") as file:
        # 3. Scrivere il contenuto nel file
        file.write(contenuto)
    print(f"Il file '{nome_file}' è stato salvato con successo!")

except IOError as e:
    print(f"Errore durante la scrittura del file: {e}")
```
